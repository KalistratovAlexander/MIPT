* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

```python
# Функция для извлечения первых двух слов
def extract_first_two_words(education_info):
    words = education_info.split(' ')
    
    # Проверка, что первое слово начинается с заглавной буквы
    if words[0][0].isupper():
        return ' '.join(words[:2])
    return None

# Создание столбца "Образование"
data['Образование'] = data['Образование и ВУЗ'].apply(extract_first_two_words)

# Удаление столбца "Образование и ВУЗ"
data_cleaned = data.drop(columns=['Образование и ВУЗ'])

# Вывод результатов
print(f"Уникальные значения в столбце 'Образование': {data['Образование'].unique()}")
data_cleaned.head()
```

2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
# Функция для извлечения возраста и пола
def parse_gender_age(value):
    # Разделяем строку по ' , '
    parts = value.split(' , ')
    # Определяем пол
    gender = 'М' if parts[0].strip() == 'Мужчина' else 'Ж'
    # Извлекаем возраст (число до слова "лет", "год" или "года")
    age = int(parts[1].split()[0])
    return gender, age

# Создание столбцов "Пол" и "Возраст"
data_cleaned[['Пол', 'Возраст']] = data_cleaned['Пол, возраст'].apply(parse_gender_age).apply(pd.Series)

# Удаление столбца "Пол, возраст"
data_cleaned.drop(columns=['Пол, возраст'], inplace=True)

# Вывод результата
data_cleaned.head()
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Списки с ключами
year_keys = ['год', 'года', 'лет']
month_keys = ['месяц', 'месяцев', 'месяца']

# Функция для извлечения опыта работы и преобразования в месяцы
def parse_experience(value):
    # Пропуски и скрытые пропуски остаются NaN
    if pd.isna(value) or value.startswith("Не указано"):
        return np.nan
    
    # Выделение первой части до описания периодов работы
    experience_part = value.split("  ")[0]
    
    years = 0
    months = 0
    
    # Проверка наличия ключевых слов для лет
    for key in year_keys:
        if key in experience_part:
            years_str = experience_part.split(key)[0].split()[-1]
            years = int(years_str)
            break
    
    # Проверка наличия ключевых слов для месяцев
    for key in month_keys:
        if key in experience_part:
            months_str = experience_part.split(key)[0].split()[-1]
            months = int(months_str)
            break 
    
    # Перевод общего опыта в месяцы
    return years * 12 + months

# Создание столбца "Опыт работы (месяц)"
data_cleaned['Опыт работы (месяц)'] = data_cleaned['Опыт работы'].apply(parse_experience)

# Удаление столбца "Опыт работы"
data_cleaned.drop(columns=['Опыт работы'], inplace=True)

# Вывод результата
data_cleaned.head()
```

* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

    * … , готов к командировкам , … 
    * … , готова к редким командировкам , …

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
# Список городов-миллионников
million_cities = [
    'Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск',
    'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь',
    'Воронеж', 'Волгоград'
]


# Списки с ключами
relocation_positive_keys = ['готов к переезду', 'готова к переезду', 'хочу переехать']
relocation_negative_keys = ['не готов к переезду', 'не готова к переезду']

travel_positive_keys = ['готов к командировкам', 'готова к командировкам', \
                        'готов к редким командировкам', 'готова к редким командировкам']
travel_negative_keys = ['не готов к командировкам', 'не готова к командировкам']

# Функция для изалечения информации о переезде и командировках
def parse_city_relocation_travel(value):
    # Инициализация значений
    city = "другие"
    relocation = False
    travel = False

    # Разделение строки на части
    parts = value.split(' , ')

    # Определяе города
    for part in parts:
        if part.strip() in ['Москва', 'Санкт-Петербург']:
            city = part.strip()
        elif part.strip() in million_cities:
            city = "город-миллионник"

    # Определение готовности к переезду
    for part in parts:
        if any(key in part for key in relocation_negative_keys): 
            relocation = False
            break
        elif any(key in part for key in relocation_positive_keys): 
            relocation = True

    # Определение готовности к командировкам
    for part in parts:
        if any(key in part for key in travel_negative_keys): 
            travel = False
            break
        elif any(key in part for key in travel_positive_keys):  
            travel = True

    return city, relocation, travel

# Создание столбцов "Город", "Готовность к переезду", "Готовность к командировкам"
data_cleaned[['Город', 'Готовность к переезду', 'Готовность к командировкам']] = \
    data_cleaned['Город, переезд, командировки'].apply(parse_city_relocation_travel).apply(pd.Series)

# Удаление столбца "Город, переезд, командировки"
data_cleaned.drop(columns=['Город, переезд, командировки'], inplace=True)

# Вывод результата
data_cleaned.head()
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Списки с ключами
employment_categories = ['полная занятость', 'частичная занятость', 'проектная работа', \
                         'волонтерство', 'стажировка']
schedule_categories = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']

# Функция для one-hot преобразования строки
def one_hot_encode(row, categories):
    return {category: category in row for category in categories}

# Создание новых столбцов с категориями занятости и графика
employment_dummies = data_cleaned['Занятость'].apply(one_hot_encode, categories=employment_categories) \
        .apply(pd.Series)
schedule_dummies = data_cleaned['График'].apply(one_hot_encode, categories=schedule_categories)\
        .apply(pd.Series)

# Добавление новых столбцов с категориями занятости и графика к исходному df
data_cleaned = pd.concat([data_cleaned, employment_dummies, schedule_dummies], axis=1)

# Удаляем столбцов "Занятость", "График"
data_cleaned.drop(columns=['Занятость', 'График'], inplace=True)

# Вывод результата
data_cleaned.head()
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Чтение и просмотр датасета с курсами валют
exchange_rates = pd.read_csv('ExchangeRates.csv')
exchange_rates.head()
```

```python
# Преобразование форматов данных
exchange_rates['date'] = pd.to_datetime(exchange_rates['date'])
exchange_rates['rate_per_unit'] = exchange_rates['close'] / exchange_rates['proportion']

# Преобразование столбца "Обновление резюме" в формат datetime и выделение даты
data_cleaned['Обновление резюме'] = pd.to_datetime(data_cleaned['Обновление резюме'], format='%d.%m.%Y %H:%M')
data_cleaned['resume_date'] = pd.to_datetime(data_cleaned['Обновление резюме'].dt.date)

# Извлечение суммы и валюты из "ЗП"
data_cleaned[['salary_amount', 'currency']] = data_cleaned['ЗП'].str.extract(r'(\d+)\s*([^\s]+)')

# Приведение наименования валюты к ISO
currency_mapping = {
    'руб.': 'RUB',
    'бел.руб.': 'BYN',
    'KZT': 'KZT',
    'EUR': 'EUR',
    'USD': 'USD',
    'грн.': 'UAH',
    'сум': 'UZS',
    'KGS': 'KGS',
    'AZN': 'AZN',
}
data_cleaned['currency_iso'] = data_cleaned['currency'].map(currency_mapping)

# Присоединение таблиц по дате и валюте
data_cleaned = pd.merge(
    data_cleaned,
    exchange_rates[['currency', 'date', 'rate_per_unit']],
    left_on=['currency_iso', 'resume_date'],
    right_on=['currency', 'date'],
    how='left'
)

# Замена курса на 1 для записей с валютой RUB
data_cleaned['rate_per_unit'] = data_cleaned['rate_per_unit'].fillna(1.0)

# Вычисление "ЗП (руб)"
data_cleaned['salary_amount'] = data_cleaned['salary_amount'].astype(float)  # Преобразуем сумму в float
data_cleaned['ЗП (руб)'] = data_cleaned['salary_amount'] * data_cleaned['rate_per_unit']

# Удаление промежуточных столбцов
data_cleaned['Обновление резюме'] = data_cleaned['resume_date']
data_cleaned = data_cleaned.drop(columns=['ЗП', 'resume_date', 'salary_amount', \
                                          'currency_x', 'currency_iso', 'currency_y', 'date', 'rate_per_unit'])

# Вывод результата
data_cleaned.head()
```

Есть ли аномалии для признака возраста, какие значения вы бы причислили к их числу?

```python
# Функция для построения гистаграммы и коробчатой диаграммы, а также для вывода статистик.
def analyze_column(data, column_name):
    # Построение графиков
    plt.figure(figsize=(14, 6))
    
    # Гистограмма
    plt.subplot(1, 2, 1)
    plt.hist(data[column_name], bins=15, edgecolor='black', alpha=0.7)
    plt.title(f'Гистограмма распределения {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Частота')
    
    # Коробчатая диаграмма
    plt.subplot(1, 2, 2)
    plt.boxplot(data[column_name], vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
    plt.title(f'Коробчатая диаграмма {column_name}')
    plt.xlabel(column_name)
    
    plt.tight_layout()
    plt.show()
    
    # Расчёт статистик
    mode = data[column_name].mode()[0]
    min_value = data[column_name].min()
    max_value = data[column_name].max()
    q1 = data[column_name].quantile(0.25)
    q3 = data[column_name].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    anomalies = data[(data[column_name] < lower_bound) | (data[column_name] > upper_bound)]
    stats = {
        "mode": mode,
        "min_value": min_value,
        "max_value": max_value,
        "bounds": (lower_bound, upper_bound),
        }
    
    return anomalies, stats
```

```python
anomalies, stats = analyze_column(data_cleaned, 'Возраст')
stats
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# Логарифмирование возраста
log_age = np.log(data_cleaned['Возраст'])

# Расчет параметров для метода трёх сигм
mean_log_age = log_age.mean()
std_log_age = log_age.std()

# Границы интервала
lower_bound = mean_log_age - 3 * std_log_age
upper_bound = mean_log_age + 4 * std_log_age

# Построение графика
plt.figure(figsize=(8, 6))
histplot = plt.hist(log_age, color='lightblue', edgecolor='black', alpha=0.7)
plt.axvline(mean_log_age, color='k', lw=2, label='Mean')
plt.axvline(lower_bound, color='r', linestyle='--', lw=2, label='Lower Bound (3σ)')
plt.axvline(upper_bound, color='g', linestyle='--', lw=2, label='Upper Bound (4σ)')
plt.title('Распределение возраста в логарифмическом масштабе')
plt.xlabel('Логарифм возраста')
plt.ylabel('Частота')
plt.legend()

# Вывод результата
plt.show()
```
